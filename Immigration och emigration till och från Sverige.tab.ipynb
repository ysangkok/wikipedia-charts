# Fetch data

```
scb.get_variables()
```

```
scb.set_query(observations=['Invandringar', 'Utvandringar'],
              ålder=["totalt"],
              kön=["totalt"], år=[str(x) for x in range(2000,2024)], födelselandgrupp=['total'])
```

Now we can check how the query looks:

```
scb.get_query()
```

```
scb.get_data()
```

```
dat = scb.get_data()
keys = [o['text'] for o in dat['columns'] if o['type'] == 'c']
keys
```

```
result = []
for o in dat['data']:
  year = int(o['key'][-1])
  immigration = int(o['values'][0])
  emigration = int(o['values'][1])
  result.append([f"{year}-12-31", immigration, emigration])

result[0:2]
```

```
import json
print(json.dumps({"license":"CC0-1.0", "schema":{"fields":[{"name":"year","type":"string"}
                                          ,{"name":"immigrations","type":"number","title":{"en":"Immigrations","sv":"Invandringar"}}
                                          ,{"name":"emigrations","type":"number","title":{"en":"Emigrations","sv":"Utvandringar"}}
                                          ]
                                }
 , "data": result
 },indent=2))
```
