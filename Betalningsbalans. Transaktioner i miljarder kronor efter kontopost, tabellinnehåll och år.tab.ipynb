# Fetch data

```
scb.get_variables()
```

Now we can check how the query looks:

```
names = scb.get_variables()['kontopost']
columncodes = scb.info()['variables'][0]['values']
name_to_code = {}
code_to_name = {}
for (name, code) in zip(names, columncodes):
  clean_name = name.partition(' ')[2]
  name_to_code[clean_name] = code
  code_to_name[code] = clean_name
name_to_code
```

```
scb.get_data()
```

```
dat = scb.get_data()
import collections
d = collections.defaultdict(list)

whitelist = ['bytesbalans, netto', 'handel i varor och tjänster', 'handelsbalans (f.o.b. - f.o.b.)', 'tjänstebalans', 'kapitalavkastning', 'sekundärinkomst']

for o in dat['data']:
  if code_to_name[o['key'][0]] not in whitelist:
    continue
  if o['values'][0] == '..':
    val = None
  else:
    val = float(o['values'][0])
  d[o['key'][1]].append(val)
d
```

```
result = []
for year, values in d.items():
  result.append([f"{year}-12-31"] + values)

result[0:2]
```

```
import json

fields = [{"name":"year","type":"string"}] + \
         [{"name":columncodes[idx].replace('.','_'),"type":"number","title":{"sv":nam.partition(' ')[2]}} for idx, nam in enumerate(names) if code_to_name[columncodes[idx]] in whitelist]
print(json.dumps({"license":"CC0-1.0", "schema":{"fields":fields}
 , "description": {
    "en": "Generated using corresponding program in https://github.com/ysangkok/wikipedia-charts"
    }
 , "data": result
 },indent=2))
```
