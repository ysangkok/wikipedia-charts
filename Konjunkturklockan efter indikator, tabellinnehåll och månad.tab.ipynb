# Fetch data

```
scb.get_variables()
```

Now we can check how the query looks:

```
names = scb.get_variables()['indikator']
columncodes = scb.info()['variables'][0]['values']
name_to_code = {}
code_to_name = {}
for (name, code) in zip(names, columncodes):
  clean_name = name.partition(' ')[2]
  name_to_code[clean_name] = code
  code_to_name[code] = clean_name
name_to_code
```

```
scb.get_data()
```

```
dat = scb.get_data()
import collections
d = collections.defaultdict(list)

allow = lambda x: True

for o in dat['data']:
  if not allow(code_to_name[o['key'][0]]):
    continue
  if o['values'][0] == '..':
    val = None
  else:
    val = float(o['values'][0])
  d[o['key'][1]].append(val)
d
```

```
result = []
import re
for mstring, values in d.items():
  month = mstring[-2:]
  year = mstring[:4]
  result.append([f"{year}-{month}-31"] + values)

result[0:2]
```

```
import json

fields = [{"name":"year","type":"string"}] + \
         [{"name":"n" + columncodes[idx].replace('.','_'),"type":"number","title":{"sv":nam.partition(' ')[2]}} for idx, nam in enumerate(names) if allow(code_to_name[columncodes[idx]])]
print(json.dumps({"license":"CC0-1.0", "schema":{"fields":fields}
 , "description": {
    "en": "Generated using corresponding program in https://github.com/ysangkok/wikipedia-charts"
    }
 , "data": result
 },indent=2))
```
